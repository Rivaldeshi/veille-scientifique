```python
import numpy as np

# Exemple de données
nombre_navettes = 5
nombre_trajets = 4
nombre_arrets = 2

# Coûts associés à l'utilisation de chaque navette pour chaque trajet à chaque arrêt
couts = np.random.randint(1, 10, size=(nombre_navettes, nombre_trajets, nombre_arrets))

# Capacité maximale des navettes pour chaque arrêt
capacites = np.random.randint(5, 15, size=(nombre_navettes, nombre_arrets))

# Temps de trajet maximal pour chaque arrêt
temps_trajet_maximal = np.random.randint(1, 5, size=(nombre_trajets, nombre_arrets))

# Disponibilité des navettes
disponibilites = np.random.choice([0, 1], size=nombre_navettes)

# Paramètres de l'algorithme de recherche tabou
taille_liste_tabou = 5
nombre_iterations = 100


# Génération d'une solution initiale valide
def generer_solution_initiale():
    solution_initiale = np.zeros((nombre_navettes, nombre_trajets, nombre_arrets))

    for i in range(nombre_navettes):
        for j in range(nombre_trajets):
            for k in range(nombre_arrets):
                # Affecter aléatoirement une navette si elle est disponible
                if disponibilites[i] == 1:
                    solution_initiale[i, j, k] = np.random.choice([0, 1])

    return solution_initiale

# Représentation de la solution
solution_initiale  = generer_solution_initiale()

# Fonction objectif
def fonction_objectif(solution):
    return np.sum(solution * couts)

# Génération de solutions voisines
def generer_solution_voisine(solution):
    nouvelle_solution = solution.copy()
    navette = np.random.randint(nombre_navettes)
    trajet = np.random.randint(nombre_trajets)
    arret = np.random.randint(nombre_arrets)
    nouvelle_solution[navette, trajet, arret] = 1 - nouvelle_solution[navette, trajet, arret]
    return nouvelle_solution

# Critère d'aspiration
def critere_aspiration(nouvelle_solution, meilleure_solution, meilleure_cout):
    nouvelle_cout = fonction_objectif(nouvelle_solution)
    return nouvelle_cout < meilleure_cout

# Algorithme de recherche tabou
def recherche_tabou(solution_initiale, taille_liste_tabou, nombre_iterations):
    meilleure_solution = solution_initiale
    meilleure_cout = fonction_objectif(meilleure_solution)
    liste_tabou = []

    for _ in range(nombre_iterations):
        voisine = generer_solution_voisine(meilleure_solution)

        if voisine.tobytes() not in liste_tabou:
            cout_voisine = fonction_objectif(voisine)

            if cout_voisine < meilleure_cout or critere_aspiration(voisine, meilleure_solution, meilleure_cout):
                meilleure_solution = voisine
                meilleure_cout = cout_voisine

            liste_tabou.append(meilleure_solution.tobytes())

            if len(liste_tabou) > taille_liste_tabou:
                liste_tabou.pop(0)

    return meilleure_solution, meilleure_cout

# Exécution de l'algorithme
meilleure_solution, meilleure_cout = recherche_tabou(solution_initiale, taille_liste_tabou, nombre_iterations)

# Affichage des résultats
print("Meilleure Solution:")
print(meilleure_solution)
print("Meilleur Coût:", meilleure_cout)
```
